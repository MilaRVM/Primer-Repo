```python
# 📦 Librerías necesarias
import fitz  # PyMuPDF
import pandas as pd
import os
import re
from datetime import datetime
from matplotlib import pyplot as plt
from IPython.display import display
# Import necessary libraries for the later steps as well, to keep it clean
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import nltk
from nltk.corpus import stopwords

# Ensure stopwords are downloaded (add quiet=True to avoid repeated messages)
nltk.download('stopwords', quiet=True)
stop_words_spanish = stopwords.words('spanish')


# 📁 Ruta donde están los PDFs (debes subir tu carpeta y descomprimirla)
# Make sure this path is correct based on your extraction step
carpeta = "/content/reforma_datos"
documentos = [] # Initialize the list to store document data

# 🧽 Limpieza de texto
def limpiar_texto(texto):
    # Keep only essential cleaning for this example, focusing on spacing
    texto = re.sub(r'\s+', ' ', texto)
    return texto.strip()

# 🔍 Palabras clave
palabras_clave = [
    "reforma parcial", "control constitucional", "enmienda", "restricción de derechos",
    "elementos constitutivos", "referéndum", "estructura fundamental", "derechos constitucionales",
    "favorable" # Make sure all keywords are here
]

# 🔍 Funciones auxiliares
def clasificar_modalidad(texto):
    texto = texto.lower()
    if "acción pública de inconstitucionalidad" in texto:
        return "Control Concentrado"
    elif "acción de protección" in texto:
        return "Control Difuso"
    elif "consulta de constitucionalidad" in texto:
        return "Control Preventivo"
    elif "hábeas corpus" in texto:
        return "Hábeas Corpus"
    elif "hábeas data" in texto:
        return "Hábeas Data"
    return "No identificado"

def extraer_fecha_archivo(nombre):
    # Example: Assuming filename is like DD-MM-YY.pdf or DD-M-YY.pdf
    # Adjust the regex if your filenames have a different format
    match = re.search(r'(\d{1,2})-(\d{1,2})-(\d{2})', nombre)
    if match:
        try:
            # Construct year assuming it's 20YY
            year = int(match.group(3))
            full_year = 2000 + year
            # Ensure month and day have leading zeros if necessary for datetime parsing
            day = int(match.group(1))
            month = int(match.group(2))

            return datetime(full_year, month, day)
        except ValueError as e:
            # Handle cases where date components are invalid (e.g., month > 12)
            print(f"Error parsing date from filename {nombre}: {e}")
            return None
        except Exception as e:
            # Catch other potential errors during date parsing
            print(f"Unexpected error parsing date from filename {nombre}: {e}")
            return None
    return None


def extraer_ponente(texto):
    # Adjusted regex to capture names which might include multiple capitalized words
    match = re.search(r'ponente\s*:\s*([A-ZÁÉÍÓÚÑ][a-záéíóúñ]+\s+[A-ZÁÉÍÓÚÑ][a-záéíóúñ]+(?:\s+[A-ZÁÉÍÓÚÑ][a-záéíóúñ]+)*)', texto, re.IGNORECASE)
    return match.group(1).strip() if match else "No identificado"


# 📄 Procesar documentos
print(f"Intentando procesar PDFs en: {carpeta}")
if os.path.exists(carpeta):
    # Use os.walk to handle subdirectories as well
    for root, dirs, files in os.walk(carpeta):
         # Skip the __MACOSX directory if it exists from extraction
        if '__MACOSX' in root:
            continue
        for archivo in files:
            if archivo.endswith(".pdf"):
                ruta = os.path.join(root, archivo)
                print(f"Procesando: {ruta}") # Print which file is being processed
                try:
                    doc = fitz.open(ruta)
                    texto = "".join(pagina.get_text() for pagina in doc)
                    texto_limpio = limpiar_texto(texto)
                    conteos = {clave: texto_limpio.lower().count(clave.lower()) for clave in palabras_clave}

                    # Check if text extraction was successful before appending
                    if texto_limpio:
                        documentos.append({
                            "archivo_pdf": archivo,
                            "ruta": ruta,
                            "n_paginas": len(doc),
                            # Reduce text length to avoid memory issues, or remove if not needed for TF-IDF
                            "texto_extraido": texto_limpio[:1500] + "..." if len(texto_limpio) > 1500 else texto_limpio,
                            "modalidad": clasificar_modalidad(texto_limpio),
                            "juez_ponente": extraer_ponente(texto_limpio),
                            "fecha_archivo": extraer_fecha_archivo(archivo),
                            **conteos
                        })
                    else:
                        print(f"Advertencia: No se pudo extraer texto limpio de {archivo}")

                except Exception as e:
                    print(f"Error procesando {archivo}: {e}")
else:
    print(f"Error: La carpeta '{carpeta}' no fue encontrada. Asegúrate de que el ZIP fue extraído correctamente.")


# 🧾 Crear DataFrame
df = pd.DataFrame(documentos)

# Check if the DataFrame was populated
if not df.empty:
    df["modalidad"] = df["modalidad"].fillna("No identificado")

    # --- Análisis y visualizaciones adicionales (from ipython-input-29) ---

    # 📊 Visualizaciones
    display(df.head(20))

    # Gráfico de modalidad
    df["modalidad"].value_counts().plot(kind='barh', title="Distribución por modalidad de control", figsize=(10,5))
    plt.gca().invert_yaxis()
    plt.show()

    # Gráfico de enmienda vs favorable
    df.plot(kind='scatter', x='enmienda', y='favorable', alpha=0.7, s=40)
    plt.title("Relación entre 'enmienda' y decisiones 'favorables'")
    plt.grid(True)
    plt.show()

    # --- TF-IDF and Clustering (from ipython-input-40) ---

    # Asegúrate de que la columna 'texto_extraido' sea de tipo string y manejar NaN
    df["texto_extraido"] = df["texto_extraido"].fillna("").astype(str)

    # Vectorizar texto
    # Ensure min_df and max_df are appropriate for your dataset size
    vectorizer = TfidfVectorizer(stop_words=stop_words_spanish, max_df=0.8, min_df=1) # min_df=1 allows single-occurrence words
    X = vectorizer.fit_transform(df["texto_extraido"]) # Use the cleaned, filled column

    # Clustering
    # Use n_init='auto' for newer sklearn versions
    kmeans = KMeans(n_clusters=5, random_state=42, n_init='auto')
    df["tema_cluster"] = kmeans.fit_predict(X)

    # Palabras clave por cluster (Themes)
    print("\n--- Temas identificados por Cluster ---")
    palabras = vectorizer.get_feature_names_out()
    cluster_themes = [] # List to store theme dictionary data

    for i in range(kmeans.n_clusters):
        centroides = kmeans.cluster_centers_[i].argsort()[::-1][:10]
        top_words = [palabras[ind] for ind in centroides]
        print(f"Tema Cluster {i}: {top_words}")
         # Store theme data in the list
        cluster_themes.append({"cluster_id": i, "top_words": ", ".join(top_words)})


    # Create DataFrame for cluster themes dictionary
    df_diccionario = pd.DataFrame(cluster_themes)

    # --- Additional Analysis and Visualizations (from ipython-input-29, ipython-input-30) ---

    # 1. Analizar la distribución de modalidades dentro de cada cluster
    print("\n--- Distribución de Modalidades por Cluster ---")
    print(df.groupby("tema_cluster")["modalidad"].value_counts().unstack(fill_value=0))

    # Visualización de la distribución de modalidades por cluster
    df.groupby("tema_cluster")["modalidad"].value_counts().unstack(fill_value=0).plot(kind='bar', figsize=(12, 7))
    plt.title("Distribución de Modalidades por Cluster Temático")
    plt.xlabel("Cluster Temático")
    plt.ylabel("Número de Documentos")
    plt.xticks(rotation=0)
    plt.legend(title="Modalidad", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # 2. Analizar la frecuencia de palabras clave dentro de cada cluster
    print("\n--- Frecuencia Promedio de Palabras Clave por Cluster ---")
    # Select only the keyword columns and the cluster column
    df_keywords_cluster = df[palabras_clave + ["tema_cluster"]]
    # Calculate the mean of the keywords by cluster
    print(df_keywords_cluster.groupby("tema_cluster")[palabras_clave].mean())

    # Visualization of keyword frequency by cluster
    df_keywords_cluster.groupby("tema_cluster")[palabras_clave].mean().plot(kind='bar', figsize=(12, 7))
    plt.title("Frecuencia Promedio de Palabras Clave por Cluster Temático")
    plt.xlabel("Cluster Temático")
    plt.ylabel("Frecuencia Promedio por Documento")
    plt.xticks(rotation=0)
    plt.legend(title="Palabra Clave", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()


    # 3. Analyze the judges associated with each cluster
    print("\n--- Jueces Ponentes por Cluster ---")
    # Show the most common judges in each cluster
    for i in range(kmeans.n_clusters):
        print(f"\nCluster Temático {i}:")
        # Ensure 'juez_ponente' is handled if there are missing values
        print(df[df["tema_cluster"] == i]["juez_ponente"].value_counts().head(5)) # Display the top 5

    # 4. Investigate specific documents from a cluster
    # You can filter the DataFrame by a specific cluster to see the documents
    print("\n--- Ejemplos de documentos en el Cluster Temático 0 ---")
    display(df[df["tema_cluster"] == 0].head()[["archivo_pdf", "modalidad", "juez_ponente", "fecha_archivo"]])


    # 5. Visualization of the relationship between 'enmienda', 'favorable', and clusters
    # You can color the scatter plot by cluster
    fig, ax = plt.subplots(figsize=(10, 6))
    scatter = ax.scatter(df['enmienda'], df['favorable'], c=df['tema_cluster'], cmap='viridis', alpha=0.7, s=40)
    ax.set_title("Relación entre 'enmienda' y 'favorables' por Cluster Temático")
    ax.set_xlabel("Conteo de 'enmienda'")
    ax.set_ylabel("Conteo de 'favorable'")
    ax.grid(True)
    # Add legend for the clusters
    legend = ax.legend(*scatter.legend_elements(), loc="upper right", title="Cluster")
    plt.show()


    # --- Saving DataFrames ---
    print("\nGuardando DataFrames...")
    # Save the main analyzed data
    df.to_csv("sentencias_analizadas.csv", index=False)
    print("DataFrame 'sentencias_analizadas.csv' guardado.")

    # Save the cluster themes dictionary
    df_diccionario.to_csv("diccionario_temas.csv", index=False)
    print("DataFrame 'diccionario_temas.csv' guardado.")

else:
    print("El DataFrame está vacío. No se pudieron procesar documentos PDF.")
```
