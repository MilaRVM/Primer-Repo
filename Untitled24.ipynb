```python
import fitz  # PyMuPDF
import pandas as pd
import os

carpeta = "reforma_datos"
documentos = []

# Palabras clave jurídicas que deseas rastrear
palabras_clave = [
    "reforma constitucional", "control previo", "acción de amparo", "constitución",
    "vacancia presidencial", "referéndum", "proceso competencial", "derechos fundamentales"
]

# Recorrer y procesar archivos PDF
for root, dirs, files in os.walk(carpeta):
    for archivo in files:
        if archivo.endswith(".pdf"):
            ruta = os.path.join(root, archivo)
            doc = fitz.open(ruta)
            texto = ""
            for pagina in doc:
                texto += pagina.get_text()
            # Contar palabras clave
            conteos = {clave: texto.lower().count(clave.lower()) for clave in palabras_clave}
            documentos.append({
                "archivo_pdf": archivo,
                "ruta": ruta,
                "n_paginas": len(doc),
                "texto_extraido": texto[:1000] + "..." if len(texto) > 1000 else texto,
                **conteos
            })

# Crear DataFrame
df = pd.DataFrame(documentos)
```

```python
# PASO 1: Extraer el ZIP si aún no se ha hecho
import zipfile, os

# Set the zip_path to the filename directly if uploaded via files.upload()
# You can verify the filename using !ls after uploading
zip_path = "Reforma constitucional.zip" # Use the correct filename if different
extract_folder = "reforma_datos"
os.makedirs(extract_folder, exist_ok=True)

# Verify the file exists before attempting to open
if not os.path.exists(zip_path):
    print(f"Error: El archivo ZIP '{zip_path}' no fue encontrado.")
    # You might want to exit or handle this error appropriately
else:
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_folder)
    print(f"✅ Archivos extraídos en: {extract_folder}")


# PASO 2: Leer PDF y crear columna 'texto'
import fitz  # PyMuPDF
import pandas as pd
import re

data = []

def limpiar(texto):
    texto = texto.replace('\n', ' ')
    texto = re.sub(r'\s+', ' ', texto)
    texto = re.sub(r'[^\w\sáéíóúñÁÉÍÓÚÑ.,:;()]', '', texto)
    return texto.strip()

for root, dirs, files in os.walk(extract_folder):
    for archivo in files:
        if archivo.endswith(".pdf"):
            ruta = os.path.join(root, archivo)
            try:
                doc = fitz.open(ruta)
                texto = "".join([pagina.get_text() for pagina in doc])
                data.append({
                    "archivo_pdf": archivo,
                    "texto": limpiar(texto)
                })
            except Exception as e:
                print(f"Error processing file {ruta}: {e}") # Add error handling for PDF processing


df = pd.DataFrame(data)
print("✅ Documentos procesados:", len(df))

# PASO 3: Buscar frases clave
frases = [
    "Corte Constitucional", "Fiscalía", "Corte Nacional", "Consejo Nacional Electoral",
    "enmienda", "asamblea constituyente", "Presidencia de la República",
    "restricción de derechos", "elementos constitutivos del estado",
    "estructura fundamental", "ciudadanía", "consulta popular"
]

for frase in frases:
    # Use a shorter column name to avoid potential issues with plotting libraries
    col = frase.lower().replace(" ", "_")[:30]
    df[col] = df["texto"].str.lower().str.contains(frase.lower()).astype(int)

# PASO 4: Grafo de coocurrencia
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

columnas = [frase.lower().replace(" ", "_")[:30] for frase in frases]
# Ensure all columns exist in the dataframe before selecting
valid_columns = [col for col in columnas if col in df.columns]
matriz = df[valid_columns].values
coocurrencia = np.dot(matriz.T, matriz)

G = nx.Graph()
# Add nodes using the valid columns
G.add_nodes_from(valid_columns)

# Adjust loop to use indices of valid_columns
for i in range(len(valid_columns)):
    for j in range(i + 1, len(valid_columns)):
        peso = coocurrencia[i, j]
        if peso > 0:
            G.add_edge(valid_columns[i], valid_columns[j], weight=peso)

# PASO 5: Visualizar red
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5, seed=42)
nx.draw_networkx_nodes(G, pos, node_size=800, node_color='lightblue')
# Ensure edges exist before attempting to draw them
if G.edges():
    nx.draw_networkx_edges(G, pos, width=[G[u][v]['weight'] for u, v in G.edges()])
nx.draw_networkx_labels(G, pos, font_size=9)
plt.title("Red de Coocurrencia de Frases Jurídicas")
plt.axis('off')
plt.show()
```
